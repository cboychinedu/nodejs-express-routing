<h4> Piping Data, Zipping and Extraction </h4> 

```javascript
/* Reading data from a file into memory */ 
// Importing the file system module 
const fs = require('fs'); 

// Creating a readStream object to read in data from 
// the file specified. With utf-8 encoding. 
const readStream = fs.createReadStream('./example.txt', 'utf8'); 
readStream.on('data', (chunk) =>
{
    // Displaying the chunk of data read into memory 
    console.log(chunk); 
})
```

```javascript
/* Zipping a file */ 
// Importing the file system module 
const fs = require('fs');  
const zlib = require('zlib'); 
const gzip = zlib.createGzip();

// Creating a constant for reading and writing the stream 
const readStream = fs.createReadStream('./example.txt'); 
const writeStream = fs.createWriteStream('.example2.tar.gz'); 

// Piping the data from the first file to the other (example2.txt) 
readStream.pipe(gzip).pipe(writeStream); 
```

<h3> Unzipping the file in Nodejs </h3> 

```javascript
// Importing the file system module 
const fs = require('fs'); 
const zlib = require('zlib'); 
const gunzip = zlib.createGunzip(); 
const readStream = fs.createReadStream('./example2.tar.gz'); 
const writeStream = fs.createWriteStream('uncompressed.txt'); 

// Piping the data 
readStream.pipe(gunzip).pipe(writeStream); 
```

```javascript
// Using routes with the http module 
const http = require('http'); 
const server = http.createServer((req, res) => 
{
    // Checking if the user is requesting for the root dir 
    if ( req.url === '/' )
    {
        // 
        res.write('Hello world from nodejs!'); 
        res.end(); 
    }
    
    // Else 
    else if ( req.url === '/home' )
    {
        // 
        res.write("This is the home route!"); 
        res.end(); 
    }
    
}); 


// Running the application 
server.listen('3000', 'localhost'); 
```

```javascript
/* Sending an HTML file */ 
// Importing the http module 
const http = require('http'); 
const fs = require('fs');  
http.createServer((req, res) => 
{
    // 
    const readStream = fs.createReadStream('./static/index.html'); 
    res.writeHead(200, {'Content-type': 'text/html'}); 
    readStream.pipe(res); 
    
    
}).listen(3000); 
```

```javascript
// Serving static file using express 
const express = require('express');  
const path = require('path'); 
const app = express(); 

// 
app.use('/public', express.static(path.join(__dirname, 'public'))); 
app.get('/', (req, res) => 
{
    // 
    res.sendFile(path.join(__dirname, 'static', 'index.html')); 
})
```
